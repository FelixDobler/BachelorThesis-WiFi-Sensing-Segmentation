```python
import csidata as cid
import seaborn as sns
import numpy as np
```

```python
data = cid.load('record/prod_data/csi/user1/2024-10-11T16-53-18-669897_seq-iw.joblib')
sel = np.index_exp[:, 2, 1, 52]
```

```python
sns.lineplot(data.csi[sel])
```

```python
diff = np.diff(data.csi[sel])
abs = np.abs(data.csi[sel])
diff_abs = np.diff(abs)
```

```python
sns.lineplot(data.csi[sel], alpha=0.5)
sns.lineplot(diff, alpha=0.5)
sns.lineplot(abs, alpha=0.5)
sns.lineplot(diff_abs, alpha=0.5)
```

```python
pos_only = np.where(data.csi[sel] < 0, np.nan, data.csi[sel])
interpol = pd.Series(pos_only).interpolate().values
sns.lineplot(interpol[10000:12000])
interpol_smooth = pd.Series(interpol).rolling(25).var().values
# interpol_smooth = (interpol_smooth - np.min(interpol_smooth)) / (np.max(interpol_smooth) - np.min(interpol_smooth))
sns.lineplot(interpol_smooth[10000:12000])
```
