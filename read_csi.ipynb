```python
import numpy as np
import pandas as pd
import seaborn as sns
import joblib
```

```python
class COMPLEX:
    def __init__(self, real, imag):
        self.real : int = real
        self.imag : int = imag

class CSI_Struct:
    def __init__(self, tstamp, channel, chanBW, rate, nr, nc, num_tones, noise, phyerr, rssi, rssi_0, rssi_1, rssi_2, payload_len, csi_len, buf_len):
        self.tstamp = tstamp
        self.channel = channel
        self.chanBW = chanBW
        self.rate = rate
        self.nr = nr
        self.nc = nc
        self.num_tones = num_tones
        self.noise = noise
        self.phyerr = phyerr
        self.rssi = rssi
        self.rssi_0 = rssi_0
        self.rssi_1 = rssi_1
        self.rssi_2 = rssi_2
        self.payload_len = payload_len
        self.csi_len = csi_len
        self.buf_len = buf_len
```

```python
# records = np.load('csi_recordings/dump_ensurance_1.npy', allow_pickle=True)
# records.shape

records = joblib.load('csi_recordings_local/philipp_pushups.joblib')
len(records[0])
```

```python
csi_matrixes = []
for record in records:
    csi = record[1]
    csi_matrixes.append(csi)
csi_matrixes = np.array(csi_matrixes)
```

```python
csi_first_antenna_pair = csi_matrixes[:, 0, 0, :]
csi_first_antenna_pair.shape
```

```python
csi_first_antenna_pair_amplitude = np.vectorize(lambda x: x.real)(csi_first_antenna_pair)
csi_first_antenna_pair_amplitude.dtype
```

```python
sns.lineplot(csi_first_antenna_pair_amplitude, legend=None)
```
