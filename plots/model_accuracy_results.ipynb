```python
import csidata as cid
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv('SegmentationModelPerformance.csv')
df = df.dropna()
df.astype({'Window Size': int, 'Test Sequence': int})
df['Accuracy'] = df['Accuracy'].str.rstrip('%').astype('float') / 100.0
```

```python
df_reduced = df.iloc[[0, 4, 11, 7]]
df_reduced.dtypes
```

```python
ax = sns.barplot(data=df_reduced, x='Model', y='Accuracy', hue='Users')
for container in ax.containers:
    ax.bar_label(container, labels=[f'{round(p.get_height()*100, 1)}%' for p in container.patches])

print(ax.containers)
```
