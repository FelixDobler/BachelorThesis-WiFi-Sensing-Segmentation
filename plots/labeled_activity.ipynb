```python
import csidata as cid
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.signal import butter, freqz, lfilter, filtfilt

plt.style.use('plots/dark.mplstyle')
```

```python
padding = 50
start, end = 601, 721
trace1 = np.index_exp[:,2, 1, 50]
trace2 = np.index_exp[:,0, 2, 49]
```

```python
csi = cid.load('csi/user1/2024-10-11T17-17-06-928945_seq-rp.raw').csi
```

```python
csi_downsampled = csi[::20, ...]
```

```python
def plotSegment(trace):
    fig, ax = plt.subplots(figsize=(6, 4))
    ax = sns.lineplot(ax=ax, data=csi_downsampled[trace])
    ax.set_xlim(start-padding, end+padding)
    ax.axvline(start, color='orange', linestyle='--')
    ax.axvline(end, color='orange', linestyle='--')
    ax.axvspan(start, end, color='orange', alpha=0.2)
    ax.set_xlabel('Frame')
    ax.set_ylabel('CSI Amplitude')
    plt.savefig(f'plots/csi_segment_{trace[1:]}.svg', bbox_inches='tight', transparent=True)
    plt.savefig(f'plots/csi_segment_{trace[1:]}.png', bbox_inches='tight', transparent=True, dpi=400)
    plt.show()
```

```python
plotSegment(trace1)
plotSegment(trace2)
```
